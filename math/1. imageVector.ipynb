```python
# colorscale 예제 색상 배열을 시각
# RGB 색상을 활용한 기본적인 이미지 생성 방법

import numpy as np
import matplotlib.pyplot as plt

arr_test = np.array(
    [
      [[0, 0, 0], [255, 0, 0], [0, 255, 0], [0, 0, 255]]
    ]
)

plt.imshow(arr_test)
plt.show()
```

```python
# 추가 학습: 2차원 배열을 확장해서 다양한 색상을 파악해보자
arr_test = np.array([
    [[0, 0, 0], [255, 0, 0], [0, 255, 0], [0, 0, 255]],  # 첫 번째 행
    [[255, 255, 0], [0, 255, 255], [255, 0, 255], [255, 255, 255]]  # 두 번째 행
])

plt.imshow(arr_test)
plt.title("2x4 RGB Colorscale")
plt.show()
```

```python
# 개별적으로 가지고 있는 이미지를 업로드 하여 그레이스케일로 저장된 벡터(행렬)을 확인해보자
# 벡터로 변환한 이미지는 다시 재생해도 데이터 손실이 없음을 확인

import cv2
import numpy as np
from google.colab import files
import os
import matplotlib.pyplot as plt

# 이미지 업로드
uploaded = files.upload()

# 업로드된 파일 이름 가져오기
filename = list(uploaded.keys())[0]
print(f"Uploaded file: {filename}")

# 그레이스케일로 이미지 불러오기
image_gray = cv2.imread(f'/content/{filename}', cv2.IMREAD_GRAYSCALE)

# 그레이스케일 이미지 확인
if image_gray is None:
    print("Error: Image not loaded. Check the filename or path.")
else:
    print("Image loaded successfully with shape:", image_gray.shape)
    plt.imshow(image_gray, cmap='gray')  # 그레이스케일 이미지 표시
    plt.title("Uploaded Gray Image")
    plt.axis('off')  # 축 제거
    plt.show()

# 이미지의 픽셀 값을 행렬 형태로 출력
# 출력할 행렬의 크기를 입력받기

try:
  number = int(input("숫자를 입력하세요: "))
  print("입력하신 숫자는:", number)
except ValueError:
  print("잘못된 입력입니다. 숫자를 입력해주세요.")

print("Image matrix (first # rows):\n", image_gray[:number, :number])

# 이미지 행렬을 .npy 형식으로 저장
np.save('image_matrix.npy', image_gray)

# 저장된 행렬 불러오기
loaded_matrix = np.load('image_matrix.npy')
print("Loaded matrix shape:", loaded_matrix.shape)

# 저장된 행렬을 다시 이미지로 저장
cv2.imwrite('restored_image.jpg', loaded_matrix)

# 두 이미지를 함께 출력
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(image_gray, cmap='gray')
axes[0].set_title("Original Gray Image")
axes[0].axis('off')

axes[1].imshow(loaded_matrix, cmap='gray')
axes[1].set_title("Restored Image from Matrix")
axes[1].axis('off')

plt.show()
```

```python

# 컬러 이미지 불러오기
image_color = cv2.imread(f'/content/{filename}', cv2.IMREAD_COLOR)
image_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)  # BGR -> RGB 변환

# 컬러 이미지 확인
if image_color is None:
    print("Error: Image not loaded. Check the filename or path.")
else:
    print("Image loaded successfully with shape:", image_rgb.shape)
    plt.imshow(image_rgb)
    plt.title("Uploaded Color Image")
    plt.axis('off')
    plt.show()

# 컬러 이미지의 픽셀 값을 행렬 형태로 출력
print("Image Color matrix (first 5 rows):\n", image_rgb[:5, :5])

# 이미지 행렬을 .npy 형식으로 저장
np.save('image_matrix_rgb.npy', image_rgb)

# 저장된 행렬 불러오기
loaded_matrix_rgb = np.load('image_matrix_rgb.npy')
print("Loaded matrix shape:", loaded_matrix_rgb.shape)

# 저장된 행렬을 다시 이미지로 저장
cv2.imwrite('restored_image_rgb.jpg', loaded_matrix_rgb)
```
