```python
import numpy as np
import matplotlib.pyplot as plt

# 가상의 데이터 생성
# X는 입력
# Y는 출력
X = np.array([1, 2, 3, 4, 5], dtype=float)
Y = np.array([10, 20, 30, 40, 50], dtype=float)

# 초기 파라미터
weight = np.random.randn()
bias = np.random.randn()
learning_rate = 0.01
epochs = 50

losses = []

# 경사 하강법 학습
for epoch in range(epochs):
    # 예측값 계산: 현재의 가중치(weight)와 편향(bias)을 사용하여 예측값을 계산
    Y_pred = weight * X + bias

    # 손실 구하기: MSE(Mean Squared Error) 사용하여 실제값(Y)과 예측값(Y_pred)의 차이를 제곱하여 평균 구하기
    loss = np.mean((Y - Y_pred) ** 2)
    losses.append(loss)

    # 그래디언트 계산: weight_gradient: 가중치에 대한 손실 함수의 기울기, bias_gradient: 편향에 대한 손실 함수의 기울기
    weight_gradient = -2 * np.mean(X * (Y - Y_pred))
    bias_gradient = -2 * np.mean(Y - Y_pred)

    # 파라미터 업데이트: 계산된 그래디언트와 학습률(learning_rate)을 사용하여 가중치와 편향을 조정
    weight -= learning_rate * weight_gradient
    bias -= learning_rate * bias_gradient

    # 10 에포크마다 현재의 손실, 가중치, 편향 값을 출력
    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.4f}, Weight = {weight:.4f}, Bias = {bias:.4f}")

# 최종 결과 출력
print(f"Final Parameters: Weight = {weight:.4f}, Bias = {bias:.4f}")

# 손실 그래프 출력
plt.plot(range(epochs), losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Reduction Over Time')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 학습률 설정
learning_rates = [0.01, 0.1, 1.0]  # 적절한 값, 너무 큰 값, 너무 작은 값
epochs = 50
loss_values = {lr: [] for lr in learning_rates}

# 초기 가중치 설정
w = 10  # 임의의 값
true_w = 2  # 실제 목표 값

for lr in learning_rates:
    w_temp = w  # 가중치 초기화
    for epoch in range(epochs):
        gradient = 2 * (w_temp - true_w)  # 미분 (손실 함수의 기울기)
        w_temp -= lr * gradient  # 가중치 업데이트
        loss = (w_temp - true_w) ** 2  # 손실 계산
        loss_values[lr].append(loss)  # 손실 저장

# 그래프 그리기
plt.figure(figsize=(8, 5))
for lr in learning_rates:
    plt.plot(loss_values[lr], label=f"Learning Rate = {lr}")

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Effect of Learning Rate on Loss Reduction")
plt.legend()
plt.grid()
plt.show()
```
