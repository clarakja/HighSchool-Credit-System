```python
# append(: 리스트 끝에 요소 추가
lst = [1, 2, 3]
lst.append(4)
print(lst)  # [1, 2, 3, 4]
```

```python
# extend: 리스트 확장(여러요소 추가)
lst = [1, 2, 3]
lst.extend([4, 5])
print(lst)  # [1, 2, 3, 4, 5]
```

```python
# insert: 특정 위치에 요소 추가
lst = [1, 2, 4]
lst.insert(2, 3) # 2번 인덱스에 3요소 추가
print(lst)  # [1, 2, 3, 4]
```

```python
# remove: 특정 위치 요소 제거
lst = [1, 2, 3, 4]
lst.remove(3) # 3번 인덱스 요소 제거
print(lst)  # [1, 2, 4]
```

```python
# sort: 리스트 정렬
lst = [4, 2, 5, 1, 3]
lst.sort()
print(lst)  # [1, 2, 3, 4, 5]
```
