```python
print("Hello, Python!")  # 문자열 출력
name = "나르셔"
print("이름:", name)  # 변수 값 출력
```

```python
name = input("이름을 입력하세요: ")
print(f"안녕하세요, {name}님!")
```

```python
x = 1
y = 2
print(f"{x} + {y}는 {x + y}입니다.")
```

```python
word = "Python"
print(f"{word}는 {len(word)}글자입니다.")
```

```python
print(f"{'left':<10}") # 왼쪽 정렬
print(f"{'center':^10}") # 가운데 정렬
print(f"{'right':>10}") # 오른쪽 정렬2
```

```python
pi = 3.141592653589
print(f'{pi:.4f}') # 소수점 4자리까지 표시
```

```python
print(int("10") + 5)   # 문자열 → 정수 변환
print(float("3.14") * 2) # 문자열 → 실수 변환
print(str(100) + "원")  # 숫자 → 문자열 변환
```

```python
age = int(input("나이를 입력하세요: "))

if age < 8:
    print("무료 탑승 가능합니다.")
elif age < 20:
    print("청소년 요금: 1000원")
elif age < 65:
    print("성인 요금: 2000원")
else:
    print("경로 우대 무료 탑승 가능합니다.")
```
