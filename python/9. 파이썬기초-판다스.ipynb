```python
import pandas as pd

data = pd.Series([10, 20, 30, 40])
print(data)
```

```python
# pd.DataFrame(): 딕셔너리 등으로 DataFrame 생성
# data 딕셔너리 생성
data = {
    "이름": ["철수", "영희", "민수"],
    "나이": [25, 23, 30],
    "도시": ["서울", "부산", "대구"]
}

df = pd.DataFrame(data)
print(df)
```

```python
# 상위 2 개 행 출력
df.head(2)
```

```python
# 특정 열/행 선택
df = pd.DataFrame({
    "이름": ["철수", "영희", "민수"],
    "나이": [25, 23, 30],
    "도시": ["서울", "부산", "대구"]
})

print(df["이름"])  # 특정 열 선택
print(df.loc[0])   # 특정 행 선택
```

```python
# '연령대' 키의 열 추가
df["연령대"] = df["나이"].apply(lambda x: "청년" if x < 30 else "중년")
print(df)
```

```python
# 나이가 25 이상인 데이터만 선택
df_filtered = df[df["나이"] >= 25]
print(df_filtered)
```

```python
df_sorted = df.sort_values(by="나이")
print(df_sorted)  # 나이 오름차순 정렬
```

```python
df_grouped = df.groupby("도시")["나이"].mean()
print(df_grouped)  # 도시별 평균 나이
```

```python
# 데이터프레임 합치기
df1 = pd.DataFrame({"이름": ["철수", "영희"], "성적": [85, 90]})
df2 = pd.DataFrame({"이름": ["철수", "영희"], "출석": [10, 8]})

df_merged = pd.merge(df1, df2, on="이름")
print(df_merged)
```

```python
df.to_csv("data.csv", index=False)  # CSV 저장
df_new = pd.read_csv("data.csv")  # CSV 불러오기
print(df_new)
```
