```python
for i in range(5):  # 0부터 4까지 반복
    print(i)
```

```python
for i in range(1, 10):  # 1부터 9까지 1씩 증가
    print(i)
```

```python
for i in range(1, 10, 2):  # 1부터 9까지 2씩 증가
    print(i)
```

```python
for char in "PYTHON":
    print(char)
```

```python
count = 1
while count <= 5:
    print(f"{count}번째 반복")
    count += 1  # 1씩 증가
```

```python
password = ""
while password != "1234":
    password = input("비밀번호를 입력하세요: ")

print("비밀번호가 맞습니다!")
```

```python
for num in range(1, 10):
    if num == 5:
        print("반복문 종료!")
        break  # 5에서 종료
    print(num)
```

```python
for num in range(1, 6):
    if num == 3:
        continue  # 3을 건너뛰고 다음 반복 실행
    print(num)
```

```python
for i in range(2, 10):  # 단(2~9)
    for j in range(1, 10):  # 곱하는 수(1~9)
        print(f"{i} x {j} = {i * j}")
    print("-" * 10)
```
