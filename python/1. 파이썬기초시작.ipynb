1.   변수 유형: 정수형, 실수형, 문자열, 값이 없는 변수
2.   변수 선언: 별도의 데이터 타입 선언 없고, 변수명 = 값 으로 선언 완료

```python
a = 10       # 정수형 변수
b = 3.14     # 실수형 변수
c = "hello"  # 문자열 변수
d = None     # 값이 없는 변수
print(type(a), type(b), type(c), type(d))  # 데이터 타입 확인
print( a, b, c, d) # 데이터 값 확인
```

```python
a, b = 10, 3
print(a + b)  # 13
print(a ** b)  # 1000 (10의 3제곱)
print(a // b)  # 3 (정수 나눗셈)
```

```python
a, b = 7, 5
print(a > b)   # True
print(a == b)  # False
print(a != b)  # True
```

```python
a, b, c = True, False, True
print(a and b)  # False
print(a or b)   # True
print(not a)    # False
```

```python
x = 10
x += 2  # x = x + 2 -> 12
x *= 3  # x = x * 3 -> 36
print(x)  # 36
```
