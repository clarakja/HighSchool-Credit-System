```python
# 초기 데이터 포함한 딕셔너리
person = {"이름": "철수", "나이": 25, "도시": "서울"}

print(person)
```

```python
person = {"이름": "철수", "나이": 25, "도시": "서울"}

print(person["이름"])  # 철수
print(person.get("나이"))  # 25
```

```python
person = {"이름": "철수", "나이": 25}

# 값 추가
person["도시"] = "서울"
print(person)  # {'이름': '철수', '나이': 25, '도시': '서울'}
```

```python
person = {"이름": "철수", "나이": 25, "도시": "서울"}

# 값 수정
person["도시"] = "강릉"
print(person)  # {'이름': '철수', '나이': 250, '도시': '강릉'}
```

```python
person = {"이름": "철수", "나이": 25, "도시": "서울"}

# 특정 키 삭제
del person["나이"]
print(person)  # {'이름': '철수', '도시': '서울'}
```

```python
person = {"이름": "철수", "나이": 25, "도시": "서울"}

# pop()으로 값 제거 및 반환
city = person.pop("도시")
print(city)  # '서울'
print(person)  # {'이름': '철수', '나이': 25}
```

```python
# dict.keys(): 모든 키(Key) 반환
person = {"이름": "철수", "나이": 25, "도시": "서울"}

keys = person.keys()
print(keys)  # dict_keys(['이름', '나이', '도시'])

# 리스트로 변환
keys_list = list(keys)
print(keys_list)  # ['이름', '나이', '도시']
```

```python
# dict.values(): 모든 값(Value) 반환
person = {"이름": "철수", "나이": 25, "도시": "서울"}

values = person.values()
print(values)  # dict_values(['철수', 25, '서울'])

# 리스트로 변환
values_list = list(values)
print(values_list)  # ['철수', 25, '서울']
```

```python
# dict.items(): (키, 값) 쌍을 튜플로 반환
person = {"이름": "철수", "나이": 25, "도시": "서울"}

items = person.items()
print(items)  # dict_items([('이름', '철수'), ('나이', 25), ('도시', '서울')])

# 리스트로 변환
items_list = list(items)
print(items_list)  # [('이름', '철수'), ('나이', 25), ('도시', '서울')]
```

```python
# dict.get(key, default): 키 조회 (없으면 기본값 반환)
person = {"이름": "철수", "나이": 25}

print(person.get("이름"))  # 철수
print(person.get("직업", "없음"))  # 기본값 없음 반환
```

```python
# sorted(): 키 또는 값 기준으로 정렬
scores = {"철수": 85, "영희": 90, "민수": 78}

# 키 기준 정렬
sorted_by_key = dict(sorted(scores.items()))
print(sorted_by_key)  # {'민수': 78, '영희': 90, '철수': 85}

# 값 기준 정렬 (오름차순)
sorted_by_value = dict(sorted(scores.items(), key=lambda x: x[1]))
print(sorted_by_value)  # {'민수': 78, '철수': 85, '영희': 90}

# 값 기준 정렬 (내림차순)
sorted_by_value_desc = dict(sorted(scores.items(), key=lambda x: x[1], reverse=True))
print(sorted_by_value_desc)  # {'영희': 90, '철수': 85, '민수': 78}
```
