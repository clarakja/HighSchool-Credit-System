```python
def say_hello():
    print("안녕하세요!")

# 함수 호출
say_hello()
```

```python
def greet(name):
    print(f"안녕하세요, {name}님!")

# 함수 호출
greet("철수")
greet("영희")
```

```python
def add(a, b):
    return a + b  # 두 수의 합 반환

# 함수 호출 후 결과 저장
result = add(3, 7)
print("합계:", result)
```

```python
def calculate(a, b):
    sum_value = a + b
    diff_value = a - b
    return sum_value, diff_value  # 두 개의 값 반환

# 함수 호출
s, d = calculate(10, 5)
print(f"합: {s}, 차이: {d}")
```

```python
# 일반 함수 (def 사용)
def add(a, b):
    return a + b

# lambda 함수로 변환
add_lambda = lambda a, b: a + b

print(add(3, 5))  # 출력: 8
print(add_lambda(3,5))
```
