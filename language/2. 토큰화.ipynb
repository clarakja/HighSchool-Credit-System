```python
# 토큰화(Tokenization)
text = "이 영화 정말 좋았어요"
tokens = text.split()
print(tokens)
# 출력: ['이', '영화', '정말', '좋았어요']
```

```python
# 파이썬의 정규 표현식(Regular Expression) 모듈을 가져오기
import re

text = "이 영화 정말 좋았어요!!!"
tokens = re.sub(r"[^\w\s]", "", text).split() #\는 ＼
print(tokens)
# 출력: ['이', '영화', '정말', '좋았어요']
```

```python
!pip install konlpy
from konlpy.tag import Okt

# OKT 한국어 형태소 분석기
okt = Okt()
text = "이 영화 정말 좋았어요"
tokens = okt.morphs(text)
print(tokens)
# 출력: ['이', '영화', '정말', '좋았어요']
```
