# !sudo apt-get install -y fonts-nanum
- 나눔 폰트를 설치
- 명령어 구성
  - sudo: 관리자 권한으로 명령어 실행
  - apt-get install [package_name]: [package_name] 패키지를 설치
  - -y: 자동 진행 옵션
    - 사용자 확인 없이 자동으로 진행
- 실행 작업
  - 시스템에 나눔 폰트 패키지를 설치
  - 설치 과정에서 필요한 의존성 패키지도 자동으로 설치


 # !sudo fc-cache -fv
- 시스템의 폰트 캐시를 강제로 갱신
- 명령어 구성
  - sudo: 관리자 권한으로 명령어 실행
  - fc-cache: 폰트 정보 캐시 파일 생성
  - -f: 강제 갱신 옵션
    - 이미 유효한 캐시가 있어도 모든 폰트 디렉토리를 다시 스캔
  - -v: 상세 출력(verbose) 옵션
    - 작업 중 상태 정보를 표시
- 실행작업
  - 시스템의 모든 폰트 디렉토리를 스캔
  - FreeType으로 읽을 수 있는 모든 폰트 파일에 대한 정보 수집
  - 각 폰트의 속성과 관련 파일명을 포함하는 캐시 생성
- 캐시를 갱신함으로써 시스템이 새로 추가된 폰트를 인식하고 사용할 수 있게 됨

#!rm ~/.cache/matplotlib -rf
- 사용자의 홈 디렉토리에 있는 Matplotlib 캐시 폴더를 강제로 삭제
- 명령어 구성
  -r 옵션: 재귀적으로 하위 디렉토리와 파일 삭제
  -f 옵션: 확인 없이 강제로 삭제 실행
- 실행작업
  - ~/.cache/matplotlib 디렉토리와 그 안의 모든 내용 제거
