```python
from sklearn.feature_extraction.text import TfidfVectorizer

# 문서 집합
documents = [
    "이 영화는 정말 재미있다",
    "이 영화는 감동적이다",
    "영화는 재미있고 감동적이다"
]

# TF-IDF 계산
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

# 단어와 TF-IDF 값 출력
print(vectorizer.get_feature_names_out())
print(tfidf_matrix.toarray())
```
